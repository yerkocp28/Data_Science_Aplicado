# Proyecto de Web Scraping Sismologia de Chile

### Importar Librerias:

```python
import urllib.request
from bs4 import BeautifulSoup
import pandas as pd
import numpy as np
from pandas.core.frame import DataFrame


e = urllib.request.urlopen("https://www.sismologia.cl/sismicidad/catalogo/2022/08/20220818.html").read()

soup = BeautifulSoup(e, 'html.parser')

# Ejemplo de como imprimir todo
# print soup.prettify()
```

```python
soup
```

```python

# Obtenemos la tabla

tabla_sismos = soup.find_all('table')[1]

# Obtenemos todas las filas
rows = tabla_sismos.find_all("tr")

```

```python
rows
```

```python

output_rows = []
for row in rows:
    # obtenemos todas las columns
    cells = row.find_all("td")
    output_row = []
    if len(cells) > 0:
        for cell in cells:
            output_row.append(cell.text)
            output_rows.append(output_row)

dataset = pd.DataFrame(output_rows)
```

```python
dataset.head()
```

```python

dataset.columns = [
    "Fecha Local / Lugar",
    "Profundidad [Km]",
    "Coordenadas"
    ,"Profundidad"
    ,"Magnitud"
]

```
